```python
#Creating the new Neurite Tables
labels_to_keep = [2,3,4,5,6,7,8]

@schema
class NeuriteContactDendriteExcitatory(dj.Computed):
    definition="""
    -> pinky.Segment
        ---
    total_n_contacts   :bigint unsigned #total number of contacts
    total_postsyn_length   :bigint unsigned #total total postsynaptic contact length
    total_n_synapses   :bigint unsigned #total total number of synapses for neurite
    apical_n_contacts   :bigint unsigned #apical number of contacts
    apical_postsyn_length   :bigint unsigned #apical total postsynaptic contact length
    apical_n_synapses   :bigint unsigned #apical total number of synapses for neurite
    basal_n_contacts   :bigint unsigned #basal number of contacts
    basal_postsyn_length   :bigint unsigned #basal total postsynaptic contact length
    basal_n_synapses   :bigint unsigned #basal total number of synapses for neurite
    oblique_n_contacts   :bigint unsigned #oblique number of contacts
    oblique_postsyn_length   :bigint unsigned #oblique total postsynaptic contact length
    oblique_n_synapses   :bigint unsigned #oblique total number of synapses for neurite
    soma_n_contacts   :bigint unsigned #soma number of contacts
    soma_postsyn_length   :bigint unsigned #soma total postsynaptic contact length
    soma_n_synapses   :bigint unsigned #soma total number of synapses for neurite
    axon_n_contacts   :bigint unsigned #axon number of contacts
    axon_postsyn_length   :bigint unsigned #axon total postsynaptic contact length
    axon_n_synapses   :bigint unsigned #axon total number of synapses for neurite
    dendrites_n_contacts   :bigint unsigned #dendrites number of contacts
    dendrites_postsyn_length   :bigint unsigned #dendrites total postsynaptic contact length
    dendrites_n_synapses   :bigint unsigned #dendrites total number of synapses for neurite
    """
    key_source = pinky.Segmentation & pinky.CurrentSegmentation
    
    def make(self,key):

        contact_pre_post = dj.U("segmentation","segment_id").aggr(
                (fc.ContactPrePost & pinky.PymeshfixDecimatedExcitatoryStitchedMesh.proj(postsyn = "segment_id")),
                    total_n_contacts="SUM(total_n_contacts)",
                    total_postsyn_length="SUM(total_postsyn_length)",
                    total_n_synapses="SUM(total_n_synapses)",
                    apical_n_contacts="SUM(apical_n_contacts)",
                    apical_postsyn_length="SUM(apical_postsyn_length)",
                    apical_n_synapses="SUM(apical_n_synapses)",
                    basal_n_contacts="SUM(basal_n_contacts)",
                    basal_postsyn_length="SUM(basal_postsyn_length)",
                    basal_n_synapses="SUM(basal_n_synapses)",
                    oblique_n_contacts="SUM(oblique_n_contacts)",
                    oblique_postsyn_length="SUM(oblique_postsyn_length)",
                    oblique_n_synapses="SUM(oblique_n_synapses)",
                    soma_n_contacts="SUM(soma_n_contacts)",
                    soma_postsyn_length="SUM(soma_postsyn_length)",
                    soma_n_synapses="SUM(soma_n_synapses)",
                    axon_n_contacts="SUM(axon_n_contacts)",
                    axon_postsyn_length="SUM(axon_postsyn_length)",
                    axon_n_synapses="SUM(axon_n_synapses)",
                    dendrites_n_contacts="SUM(dendrites_n_contacts)",
                    dendrites_postsyn_length="SUM(dendrites_postsyn_length)",
                    dendrites_n_synapses="SUM(dendrites_n_synapses)")



        #print(contact_pre_post)
        self.insert(contact_pre_post,skip_duplicates=True)
            
```

```python
""" Without dividing them up into excitatory and inhibitory"""
compartment_names = ["Apical","Basal","Oblique","Dendrite"]
# Try to get the total contact conversion in the volume overall and for these neurons (Excitatory)
total_neurite_synapses = fc.NeuriteContact.proj("apical_n_synapses",
                                                                 "basal_n_synapses",
                                                                 "oblique_n_synapses",
                                                                 "dendrites_n_synapses")
total_neurite_contacts = fc.NeuriteContact.proj("apical_postsyn_length",
                        "basal_postsyn_length",
                       "oblique_postsyn_length",
                        "dendrites_postsyn_length"
                       )

#fetch the total apical, basal,dendrite and oblique synapse number
total_synapses = dj.U("segmentation").aggr(fc.NeuriteContact,
                          apical_num="SUM(apical_n_synapses)",
                          basal_num="SUM(basal_n_synapses)",
                          oblique_num="SUM(oblique_n_synapses)",
                          dendrite_num="SUM(dendrites_n_synapses)")

total_synapse_count = total_synapses.fetch()
total_synapse_count = np.array(list(total_synapse_count[0]))[1:]
total_synapse_count = total_synapse_count.astype("int")
print_stats(compartment_names," Total Excitatory Synapses in Volume",total_synapse_count)

print("---------------------------------------------\n")

#fetch the total apical, basal,dendrite and oblique contact length
total_contacts = dj.U("segmentation").aggr(fc.NeuriteContact,
                          apical_num="SUM(apical_postsyn_length)",
                          basal_num="SUM(basal_postsyn_length)",
                          oblique_num="SUM(oblique_postsyn_length)",
                          dendrite_num="SUM(dendrites_postsyn_length)")

total_contacts_count = total_contacts.fetch()
total_contacts_count = np.array(list(total_contacts_count[0]))[1:]
total_contacts_count = total_contacts_count.astype("float64")
print_stats(compartment_names," Total Excitatory Contact Length in Volume",total_contacts_count)

print("---------------------------------------------\n")

#get the overall contact conversion and the contact conversion for compartments
total_contact_conversion = np.sum(total_synapse_count)/np.sum(total_contacts_count)
print("total_contact_conversion = %f" % (total_contact_conversion))
compartment_names = ["Apical","Basal","Oblique","Dendrite"]
for i,comp_name in enumerate(compartment_names):
    print(f"{comp_name} contact conversion = %f" % (total_synapse_count[i]/total_contacts_count[i]))
```
