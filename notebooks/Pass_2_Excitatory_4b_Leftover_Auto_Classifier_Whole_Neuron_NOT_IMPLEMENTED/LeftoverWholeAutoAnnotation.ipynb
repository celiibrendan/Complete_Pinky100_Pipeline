```python
@schema
class LeftoverWholeAutoAnnotations(dj.Computed):
    definition = """
    # creates the labels for the mesh table
    -> pinky.ExcitatoryLeftoverMeshes
    ---
    vertices   : longblob     # label data for the vertices
    triangles  : longblob     # label data for the faces
    """
    
    def make(self, key):
        start_time = time.time()
        
        print("\n\n*****Starting Auto Labeling for " + str(key["segment_id"]) + "******")
        
        segment_id = key["segment_id"]

        n_verts,n_faces = (pinky.ExcitatoryLeftoverMeshes() & key).fetch1("n_vertices","n_triangles")
        n_verts,n_faces
        verts_labels = np.zeros(n_verts)
        faces_labels = np.zeros(n_faces)
        
        #insert the key
        self.insert1(dict(key,
                  vertices=verts_labels,
                  triangles=faces_labels),skip_duplicates=True)


        
        print(f"Finished Auto Labeling: {time.time() - start_time}")
        
```

```python
start = time.time()
LeftoverWholeAutoAnnotations.populate(reserve_jobs=True)
print(time.time() - start)
```
