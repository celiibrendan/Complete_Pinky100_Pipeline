```python
"""
Pseudocode for juxtaposition
1) Create a dictionary for each face with a list
2) iterat through all the faces:
    access the dictionary key of that face
    increment the label that was actually given it
3) Print a grid of the percentage distributions


"""
```

```python
accuracy_dict = dict([(k,[0]*12) for k in range(2,13)])
accuracy_dict
```

```python
#check that they have the same number of keys
if len(auto_classifier_data.keys()) != len(proofread_data.keys()):
    raise Exception("length of keys don't match")
```

```python
from tqdm import tqdm
import time
start_time  = time.time()
for proof_key in tqdm(proofread_data.keys()):
    #get the proofread data
    proof_labels = proofread_data[proof_key]
    #get the automated labels
    auto_labels = auto_classifier_data[proof_key]

    #iterate through the proof_labels and increment the ones that were the same
    for a,p in zip(auto_labels,proof_labels):
        if a >= 13:
            a = 13
        accuracy_dict[int(p)][int(a-2)] =  accuracy_dict[int(p)][int(a-2)] + 1

print(time.time() - start_time)
```

```python
labels_names = pinky.LabelKey.fetch("description") #& "numeric< 13" #& "numeric >=2"
total_names = labels_names[2:13]
```

```python
list(total_names) + ["unclassified"]
```

```python
correct_name_dict = dict()

for i,name in enumerate(total_names):
    correct_name_dict[name] = list(np.array(accuracy_dict[i+2])/sum(accuracy_dict[i+2]))
    
correct_name_dict["name"] = list(total_names) + ["unclassified"]
correct_name_dict
```

```python
import pandas as pd
pf = pd.DataFrame(correct_name_dict)#.set_index("name",inplace=True)
pf.set_index("name")
```
