```python
labels_dict = dict(
spine_1=12,
    shaft=2,
    shaft_over_spined=18,
    shaft_3=5,
    remove_label=0,
    error=10,
    spine_2=3,
    spine_3=4,
    spine_4=6,
    spine_5=9,
    all_one_spine=11,
    auto_shaft=17,
    auto_spine=14,
    spine_head_manual=13,
    spine_neck_manual=15,
    label_removed=1
)

labels_dict_flipped = dict([(v,k) for k,v in labels_dict.items()])
labels_dict_flipped
```

```python
"""
Going to download all of the labels and see what the distributions are
"""
from collections import Counter
search_key = dict(author_proofreader="js48",status="complete")
complete_table = pinky.ProofreadLabelSpineWithInterestSlices() & search_key
vert_labels,segment_id = complete_table.fetch("vertices","segment_id")

total_Counter = Counter()

for seg_id,lab in zip(segment_id,vert_labels):
    current_counter = Counter(lab)
    if 1 in current_counter.keys():
        print(f"{seg_id} has label 1")
    if 11 in current_counter.keys():
        print(f"{seg_id} has label 10")
    if 15 in current_counter.keys():
        print(f"{seg_id} has label 15")
    total_Counter += current_counter

total_Counter
# map the number to the actual labels
label_name_counts = dict([(labels_dict_flipped[k],total_Counter[k]) for k in dict(total_Counter).keys() if k in labels_dict_flipped.keys() ])
label_name_counts
```
