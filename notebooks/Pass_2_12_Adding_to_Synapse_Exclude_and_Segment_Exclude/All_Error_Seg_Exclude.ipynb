# Finding all the errored segments in Excitatory

```python
pinky.CompartmentFinal.ComponentFinal() & "segment_id=648518346349475534"
```

```python
excitatory_seg_id,excitatory_type = pinky.CompartmentFinal.ComponentFinal.fetch("segment_id","compartment_type")
len(np.unique(excitatory_seg_id))

excitatory_seg_types = dict()
excitatory_seg_types_unique = dict()
for seg in excitatory_seg_id:
    excitatory_seg_types[seg] = []
    excitatory_seg_types_unique[seg] = []

for seg,comp_type in zip(excitatory_seg_id,excitatory_type):
    excitatory_seg_types[seg].append(comp_type)
    
#print(excitatory_seg_types)
#np.unique(excitatory_seg_types[648518346341371119])
#print(len(np.unique(excitatory_seg_types[648518346341371119])))
#print(str(np.unique(excitatory_seg_types[648518346341371119])[0]))


#only get the unique types
all_error_segments = []
for key,value in excitatory_seg_types.items():
    unique_values = np.unique(value)
    if len(unique_values) == 1:
        if str(unique_values[0]) == "Error":
            all_error_segments.append(key)

print(all_error_segments)





```

```python
pinky.SegmentExclude()
```

```python
#write them to the database

#assemble the dictionary
excitatory_segment_excludes = [dict(segmentation=3,segment_id=k,criteria_id=2) for k in all_error_segments]
excitatory_segment_excludes

pinky.SegmentExclude.insert(excitatory_segment_excludes,skip_duplicates=True)


```
