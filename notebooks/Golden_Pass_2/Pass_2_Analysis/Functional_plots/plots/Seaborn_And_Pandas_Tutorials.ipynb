# Seaborn different types of plots

```python
#catplot
"""
https://seaborn.pydata.org/generated/seaborn.catplot.html

Observations: 
1) x and y tell the columns to use for that axis
2) hue: tells what color categories to make each of the data points
3) data just points your DataFrame
4) kind = tells what kind of plot to have
5) col = Create another categorical variables by that grouping which creates different plot
6) col_wrap = 4 : When have multiple column facets, can wrap them around using this number
--> so if had 6 graphs then would wrap around after the 4th and start on next row
"""

import seaborn as sns
sns.set(style="ticks")
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise)
g
```

```python
"""
Can get the size of each of the groups if not restricted
"""
import pandas as pd
import numpy as np

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

grouped = df.groupby('Team')
print(grouped.agg(np.size))

#can apply multiple calculations to the group and will all be new columns
print(grouped['Points'].agg([np.sum, np.mean, np.std]))
new_table = grouped['Points'].agg([np.sum, np.mean, np.std])
new_table
```

```python
"""
Transformations with Pandas
- returns object that is indexed the same size as that being grouped

Conclusion: Applies the function to each of the groups and then 

Makes sense now: what it does
1) Goes and groups everything by team
2) For each team: 
    a. Takes that subset of the rank for only that team
    b. Sends it to the score function
    c. The function will either send back an array or just one number
    d. If just sends back one number then copies and pastes it in the possition of every
        row that was in that subset

"""

# import the pandas library
import pandas as pd
import numpy as np

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

grouped = df.groupby('Team')
score = lambda x: (x - x.mean()) / x.std()*10
grouped["Rank"].transform(score)
```

# add text and values on the heatmap

```python
data= np.random.rand(4,6)
text = np.asarray([['a', 'b', 'c', 'd', 'e', 'f'],
                   ['g', 'h', 'i', 'j', 'k', 'l'], 
                   ['m', 'n', 'o', 'p', 'q', 'r'], 
                   ['s', 't', 'u', 'v', 'w', 'x']])
labels = (np.asarray(["{0}\n{1:.2f}".format(text,data) for text, data in zip(text.flatten(), data.flatten())])).reshape(4,6)
"""
the 0 is the text and the 1:.2 formats the float
"{0}\n{1:.2f}"
"""

labels
```

```python
"""
How to apply the labels to the parts of the heatmap
"""
sb.set(font_scale=1.2) #can work above or below
heat_map = sb.heatmap(data,annot=labels, fmt="")
```
