```python
import time
import numpy as np
@schema
class DatajointExampleXaqAnalysis(dj.Computed):
    definition = """
    # creates the labels for the mesh table
    -> ta3p100.DatajointExampleXaq
    ---
    sum_of_weights   : decimal(10,3)     # sum of all their weights
    """
    
    def make(self,key):
        start_time = time.time()
        print("working on " + str(key))
        data = (ta3p100.DatajointExampleXaq & key).fetch1()
        network_values = data["network_values"]
        
        sum_network = np.sum(network_values)
        time.sleep(60)
        
        self.insert1(dict(key,sum_network))
        print(f"Total time = {time.time() - start_time}")
        
```

```python
DatajointExampleXaqAnalysis.populate(reserve_jobs=True)
```
